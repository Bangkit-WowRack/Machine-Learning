```python
import pandas as pd
df = pd.read_csv('random_cpu_usage7.csv')
df.head()
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
x = df[['cpu_usage', 'memory_usage']].values
```

```python
k_means = KMeans(n_clusters=3, random_state=1, n_init=1).fit(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:, 0], x[:, 1], c=k_means.labels_)
plt.xlabel('CPU Usage')
plt.ylabel('Memory Usage')
plt.show()
```

```python
#time series anomaly detection using RNN
 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, cohen_kappa_score
```

```python
# read data
df = pd.read_csv('../../data/m16_cpu_usage_v2_10minutes_anomaly.csv')
```

```python
df.head()
```

```python
# data preprocessing
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')
df = df.sort_index()
```

```python
df['cpu_util'] = (df['cpu_util']/100).astype(float)
```

```python
X = df['cpu_util'].values
y = df['anomaly'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
X_train
```

```python
X_train.shape
```

```python
# data normalization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, 1))
X_test = scaler.transform(X_test.reshape(-1, 1))
```

```python
# data reshape
X_train = X_train.reshape(-1, 1, 1)
X_test = X_test.reshape(-1, 1, 1)

X_train.shape
```

```python
# build model
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(128, input_shape=(1, 1), return_sequences=True))
model.add(tf.keras.layers.LSTM(64))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# train model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

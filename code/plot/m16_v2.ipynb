```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/m16_cpu_usage_v2.csv', delimiter=';' )
```

```python
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
df.head()
```

```python
df['cpu_util'] = df['cpu_util'].astype(float)
```

```python
df.head()
```

```python
# skip data every 10 minutes
df = df.iloc[::10, :]
```

```python
df.head()
```

```python
#plot
plt.figure(figsize=(20,10))
plt.plot(df['time'], df['cpu_util'])
plt.xlabel('time')
plt.ylabel('cpu_util')
plt.show()
```

```python
#save to csv
df.to_csv('../data/m16_cpu_usage_v2_10minutes.csv', index=False)
```

```python
#plot from 2017-10-05 to 2017-10-06
df = pd.read_csv('../data/m16_cpu_usage_v2_10minutes.csv', delimiter=',' )
df['time'] = pd.to_datetime(df['time'])
d = df[(df['time'] >= '2017-10-05') & (df['time'] <= '2017-10-13')]
plt.figure(figsize=(20,10))
plt.plot(d['time'], d['cpu_util'])
plt.xlabel('time')
plt.ylabel('cpu_util')
plt.show()
```

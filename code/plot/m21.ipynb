```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/m21_mem_usage.csv', delimiter=';')
```

```python
data.head()
```

```python
data['mem_usage'] = ((data['mem_total'] - data['mem_free'])/data['mem_total']).astype(float)
```

```python
data.head()
```

```python
data['time'] = pd.to_datetime(data['time'])
```

```python
data.dropna()
```

```python
 #plot
plt.figure(figsize=(20,10))
plt.plot(data['time'], data['mem_usage'])
plt.xlabel('Time')
plt.ylabel('Memory Usage')
plt.show()
```

```python
#skip data every 10 minutes
data = data.iloc[::10, :]
data
```

```python
 #plot
plt.figure(figsize=(20,10))
plt.plot(data['time'], data['mem_usage'])
plt.xlabel('Time')
plt.ylabel('Memory Usage')
plt.show()
```

```python
#save data
data.to_csv('../data/m21_mem_usage_clean.csv', index=False)
```

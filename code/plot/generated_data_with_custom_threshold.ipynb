```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../output4.csv')
```

```python
df.head()
```

```python
df.set_index('id')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df.sort_values(by=['date'])
```

```python
cpu_data = df[['cpu_util', 'date']].sort_values(by=['date'])
```

```python
cpu_data
```

```python
#plot cpu_data
plt.figure(figsize=(20,10))
plt.plot(cpu_data['date'], cpu_data['cpu_util'])
plt.xlabel('Date')
plt.ylabel('CPU Utilization')
#from 2023-07 to 2023-08
plt.xlim(pd.Timestamp('2023-07-25'), pd.Timestamp('2023-08-30'))
plt.show()
```

```python

#get from 2023-07 to 2023-08
cpu_data_2 = cpu_data[(cpu_data['date'] > '2023-07-25') & (cpu_data['date'] < '2023-08-30')]
```

```python
cpu_data_2.__len__
```

```python
cpu_data_2.to_csv('../generated_cpu_data_test_2.csv', index=False)
```

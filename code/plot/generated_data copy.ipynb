```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../output4.csv')
```

```python
df.head()
```

```python
df.set_index('id')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df.sort_values(by=['date'])
```

```python
cpu_data = df[['cpu_util', 'date']].sort_values(by=['date'])
```

```python
cpu_data
```

```python
#plot cpu_data
plt.figure(figsize=(20,10))
plt.plot(cpu_data['date'], cpu_data['cpu_util'])
plt.xlabel('Date')
plt.ylabel('CPU Utilization')
#from 2023-07 to 2023-08
plt.xlim(pd.Timestamp('2023-07-25'), pd.Timestamp('2023-08-30'))
plt.show()
```

```python
memory_data = df[['mem_used', 'mem_specs', 'date']].sort_values(by=['date'])
```

```python
memory_data['mem_used'] = memory_data['mem_used']/memory_data['mem_specs']
```

```python
memory_data
```

```python
#plot cpu_data
plt.figure(figsize=(20,10))
plt.plot(memory_data['date'], memory_data['mem_used'])
plt.xlabel('Date')
plt.ylabel('Memory Utilization')
#from 2023-07 to 2023-08
plt.xlim(pd.Timestamp('2023-07-25'), pd.Timestamp('2023-09-25'))
plt.show()
```

```python
# get data from 2023-07-25 to 2023-08-10

memory_data_2 = memory_data[(memory_data['date'] > '2023-07-25') & (memory_data['date'] < '2023-09-25')]
```
